```python
#used to train a model and split the tarining,evaluation data
from sklearn.model_selection import train_test_split
#it split data in train and test.
#10 for training 20 for testing else for validation
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5, shuffle = True) 
```

```python
#knn(k nearest neighour)
#k means number of neibours
#k depend k hum os ki value or neiboghers ketnay select krtyhai.
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(5)
```

```python
knn.fit(x_train,y_train)
```

```python
predictions = knn.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy1 = accuracy_score( predictions,y_test)
```
